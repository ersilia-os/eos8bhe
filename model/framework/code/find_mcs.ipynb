```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import rdFMCS

def find_common_substructure(molecules):
    # Convert SMILES strings to RDKit molecules
    mol_list = [Chem.MolFromSmiles(smi) for smi in molecules]
    # Use the FindMCS function to find the Maximum Common Substructure
    mcs = rdFMCS.FindMCS(mol_list)
    # Extract the common substructure from the first molecule
    common_substructure = Chem.MolFromSmarts(mcs.smartsString)
    return common_substructure
```

```python
# Example with a subset of molecules from the OSM dataset
subset_smiles = [
    "N#Cc1ccc(-c2nnc3cncc(OCCc4ccc(F)c(F)c4)n23)cc1",
    "FC(F)Oc1ccc(-c2nnc3cncc(OCc4cn(-c5ccccc5)nn4)n23)cc1",
    "OCC(COc1cncc2nnc(-c3ccc(OC(F)F)cc3)n12)c1cccc(OCc2ccccc2)c1",
    "FC1=CC=C(CCOC2=CN=CC3=NN=C(N23)C2=CC=C(C(OC3CC3)=C2)C2=CC(OC3CC3)=C(Cl)C=C2)C=C1F",
]

common_substructure = find_common_substructure(subset_smiles)
print(Chem.MolToSmiles(common_substructure))

# Visualize the common substructure
Chem.Draw.MolToImage(common_substructure)
```
